```python
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('/content/flower_image.jpeg')
cv2_imshow(img)
```

```python
gray_scale_image=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)        # Converting RGB image to Gray scale image
cv2_imshow(gray_scale_image)
```

```python
negative_image=255-gray_scale_image         # Converting grey scale image to negative image
cv2_imshow(negative_image)
```

```python
blurred_image=cv2.GaussianBlur(negative_image,(25,25),0)              #Blurring the negative image
cv2_imshow(blurred_image)
```

```python
negative_image=255-blurred_image                        # Pencil sketch achieved by combining the blurry negative image with gray scale image
result_sketch=cv2.divide(gray_scale_image, negative_image,scale=256.0)
cv2_imshow(result_sketch)
```

```python
fig=plt.figure(figsize=(10,9))

fig.add_subplot(1,2,1)
plt.imshow(img)
plt.axis('off')
plt.title('Input Image')

fig.add_subplot(1,2,2)
plt.imshow(result_sketch,cmap="gray")
plt.axis('off')
plt.title('Pencil Sketch')
```
